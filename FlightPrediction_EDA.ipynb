## Flight price prediction(EDA + feature engineering)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train = pd.read_excel('Data_Train.xlsx')
df_test = pd.read_excel('Test_set.xlsx')
```

```python
df = df_train.append(df_test)
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
# using lamba function

#df['date'] = df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
#df['Month'] = df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
#df['Year'] = df['Date_of_Journey'].apply(lambda x:x.split('/')[2])
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.drop(['Date_of_Journey'],axis=1, inplace= True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
```

```python
#df['Arrival_time'] = df['Arrival_time'].apply(lambda x:x.split(' ')[0])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Arrival_hour'] = df['Arrival_Time'].apply(lambda x:x.split(':')[0]).astype(int)
df['Arrival_min'] = df['Arrival_Time'].apply(lambda x:x.split(':')[1]).astype(int)
df.drop(['Arrival_Time'],axis=1, inplace= True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Dep_Hour'] = df['Dep_Time'].apply(lambda x:x.split(':')[0]).astype(int)
df['Dep_Min'] = df['Dep_Time'].apply(lambda x:x.split(':')[1]).astype(int)
df.drop(['Dep_Time'],axis=1, inplace= True)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop' : 0, '1 stop' : 1, '2 stops' : 2, '3 stops' : 3, '4 stops' : 4, 'nan':2})
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
df['Airline'] = labelencoder.fit_transform(df['Airline'])
df['Source'] = labelencoder.fit_transform(df['Source'])
df['Destination'] = labelencoder.fit_transform(df['Destination'])
df['Additional_Info'] = labelencoder.fit_transform(df['Additional_Info'])
```
